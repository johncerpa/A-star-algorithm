```python
from PIL import Image
import pytesseract
import re
from collections import defaultdict
from pprint import pprint
from a_estrella import Graph
import networkx as nx 
import numpy as np 
```

```python
nombreImagen = 'matrix2.png'
cadena = pytesseract.image_to_string(Image.open(nombreImagen), config='digits').strip()

# Construir la matriz a partir de la cadenas
matrizAdy = []
for linea in cadena.splitlines():
    listaNums = linea.split(' ')
    fila = []
    for i in listaNums:
        fila.append(int(re.sub("[^0-9]", "", i))) # Solo numeros
    matrizAdy.append(fila)

listaAdy = defaultdict(list)
edges = set()

# Construir lista de adyacencia dada la matriz de ady.
for i, v in enumerate(matrizAdy, 0):
    for j, u in enumerate(v, 0):
        if u != 0 and frozenset([i, j]) not in edges:
            edges.add(frozenset([i, j]))
            listaAdy[i].append((j, u))

myGraph = Graph(listaAdy)
myGraph.a_star_algorithm(0, 9)

G = nx.from_numpy_matrix(np.array(matrizAdy))  
nx.draw(G, with_labels=True) 
```
